```python
tinggi = [log11.loc[0]["AverageRTT(ms)"],log12.loc[0]["AverageRTT(ms)"], log13.loc[0]["AverageRTT(ms)"]]
datar = [1,2, 3]

plt.xlabel("Content Size")
tick_label = ["30","60","90"]

plt.bar(datar, tinggi, tick_label = tick_label, width = 0.5)

plt.ylabel("Average RTT")
plt.title("Content Size vs RTT")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Gabungkan DataFrames df1, df2, dan df3
combined_df1 = pd.concat([log11[['PatternID', 'AverageRTT(ms)']], log12[['PatternID', 'AverageRTT(ms)']], log13[['PatternID', 'AverageRTT(ms)']]], ignore_index=True)

# Membuat line chart dengan warna yang berbeda-beda
plt.figure(figsize=(8, 6))  # Ukuran figur (opsional)

# Daftar warna yang berbeda untuk setiap DataFrame
colors = ['blue', 'green', 'red']

for i, df in enumerate([log11, log12, log13]):
    plt.plot(df['PatternID'], df['AverageRTT(ms)'], marker='o', linestyle='-', color=colors[i], label=f'DF{i+1}')

plt.title('Grafik Line Chart: Waktu vs. TotalRTT(ms)')
plt.xlabel('PatternID')
plt.ylabel('RTT')
plt.grid(True)  # Menampilkan grid (opsional)
plt.legend()  # Menampilkan legenda untuk setiap line
plt.show()
```
