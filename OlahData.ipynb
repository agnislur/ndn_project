```python
log11=pd.read_csv("log-skenario1-1.csv")
log12=pd.read_csv("log-skenario1-2.csv")
log13=pd.read_csv("log-skenario1-3.csv")
log21=pd.read_csv("log-skenario2-1.csv")
log23=pd.read_csv("log-skenario2-3.csv")
log31=pd.read_csv("log-skenario3-1.csv")
log32=pd.read_csv("log-skenario3-2.csv")
log33=pd.read_csv("log-skenario3-3.csv")
```

```python
import pandas as pd

gabunglog1 = pd.concat([log11["TotalRTT(ms)"], log12["TotalRTT(ms)"], log13["TotalRTT(ms)"]], axis=0, ignore_index=True)
gabunglog1
```

```python
import matplotlib.pyplot as plt

# Mengurutkan data umur jika diperlukan
gabunglog1 = gabunglog1.sort_values()

# Menghitung jumlah data untuk setiap umur
umur_counts = gabunglog1.value_counts().sort_index()

# Membuat line chart
plt.plot(umur_counts.index, umur_counts.values, marker='o', linestyle='-', color='blue')
plt.title('Distribusi Umur')
plt.xlabel('Umur')
plt.ylabel('Jumlah')
plt.grid(True)  # Menampilkan grid
plt.show()
```

```python
import pandas as pd

# Tambahkan kolom baru dengan urutan 1, 2, 3, ...
log11['Urutan'] = range(1, len(log11) + 1)
log12['Urutan'] = range(1, len(log12) + 1)
log13['Urutan'] = range(1, len(log13) + 1)

# Tampilkan DataFrame hasil
print(log11)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Gabungkan DataFrames df1, df2, dan df3
combined_df1 = pd.concat([log11[['Urutan', 'InterestSent']], log12[['Urutan', 'InterestSent']], log13[['Urutan', 'InterestSent']]], ignore_index=True)

# Membuat line chart dengan warna yang berbeda-beda
plt.figure(figsize=(8, 6))  # Ukuran figur (opsional)

# Daftar warna yang berbeda untuk setiap DataFrame
colors = ['blue', 'green', 'red']

for i, df in enumerate([log11, log12, log13]):
    plt.plot(df['Urutan'], df['InterestSent'], marker='o', linestyle='-', color=colors[i], label=f'DF{i+1}')

plt.title('Grafik Line Chart: Waktu vs. InterestSent')
plt.xlabel('Urutan')
plt.ylabel('RTT')
plt.grid(True)  # Menampilkan grid (opsional)
plt.legend()  # Menampilkan legenda untuk setiap line
plt.show()
```
